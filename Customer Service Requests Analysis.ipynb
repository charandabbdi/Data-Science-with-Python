# Statistical Test

## 1.) Whether the average response time across complaint types is similar or not (overall)
### - Null hypothesis (h0): Average response time across complaint types is similar.
### - Alternate hypothesis (h1): Average response time across complaint types is not similar.

```python
mean_ser=Complaint_grp['Request_Closing_Time(in hours)'].mean().round(2)
mean_ser
```

```python
import scipy.stats as stats
z_score=stats.zscore(mean_ser,nan_policy='omit').round(3)
z_score
```

```python
P_value = stats.norm.sf(abs(z_score)).round(5)
P_value
```

## Let's take three random 'z_score' values and find the 'p_values' for them. Let's take the "significance level(alpha)=0.05"

```python
P_value_1 = stats.norm.sf(abs(z_score[2]))
P_value_2 = stats.norm.sf(abs(z_score[7]))
P_value_3 = stats.norm.sf(abs(z_score[16]))
print(f'p-value for z-score {z_score[2]} is : {P_value_1:.7f}\np-value for z-score {z_score[11]} is : {P_value_2:.4f}\np-value for z-score {z_score[22]} is : {P_value_3:.4f}')
```

## Conclusion : As we can see that "p-value" for any random "z-score" is greater than the "significance level(alpha)" except the one at 3rd index i.e "z_score[2]" in z_score list which is less than "significance level(alpha)" 
## Reject the null hypothesis (h0)

#  Statistical Test
## 2.) Are the type of complaint or service requested and location related?
## Null hypothesis (h0): Complaint Type and Location(City) are independent
## Alternate hypothesis (h1): Complaint Type and Location(City) are not independent

```python
r_df['City']=r_df['City'].str.upper()
cross_tab=pd.crosstab(r_df['Complaint Type'],r_df['City'])
cross_tab
```

```python
Chi_val,p_val,dof,exp_val=stats.chi2_contingency(cross_tab)
print(f'Chi-square Value: {Chi_val}\nP-value: {p_val}\nDegree of Freedom: {dof}\nExpected Value: {exp_val:}')
```

## Conclusion: Reject the null hypothesis (h0)
